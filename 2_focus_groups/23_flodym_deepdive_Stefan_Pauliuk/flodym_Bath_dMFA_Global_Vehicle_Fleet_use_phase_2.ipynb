# Estimating the material content of the global vehicle fleet - step 2: stock-driven model for the use phase
*derived from flodym example 5*

This example shows a stock-driven model to estimate the passenger vehicle fleet, covering multiple countries and the years 1993-2060.

The research questions asked is: How many newly registered vehicles are there in the different countries based on the stock-driven model?

To answer these questions a dynamic material flow analysis is performed.

The dynamic MFA model has the following indices:

* t: time (1993-2024)
* r: region (39 countries for which inflow data are easily available)
* g: good (passenger vehicle) (only for the second part)
