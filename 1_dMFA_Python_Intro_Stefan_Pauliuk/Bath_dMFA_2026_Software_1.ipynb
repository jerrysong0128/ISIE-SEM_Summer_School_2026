### Task 2: Calculate the number of electric vehicles registered so far. Plot and save the result.

For each age-cohort *c*, the time series *I(c)* that we read above indicates the total number of newly registered vehicles. 

The _BEVshare(c,T)_ parameter indicates the share of conventional (first column _T=0_) and electric (second column _T=1_) passenger vehicles in the new registration.

The split of the total inflow into different techology types, *I(c,T)*, is given by the following equation:

$$ I(c,T) = I(c) \cdot BEVshare(c,T) /100  $$

(Yes, you can insert equations in \LaTeX formatting into the markdown fields in the Jupyter notebooks!)

The factor of 1/100 is multiplied to go convert the unit from percent-values to share (unit 1). Below, we first read the data for *BEVshare(c,T)* into the workspace and then compute the split inflow, using the _np.einsum_ function for multiplication and summation across multiple indices.

```python
df_BEVshare = pd.read_excel('Bath_dMFA_2026_Software_Data.xlsx', sheet_name='BEV_share', index_col=0)
df_BEVshare # print data
```

```python
I_c          = df_I.values[:,0]   # convert pandas dataframe values to numpy array. Slice out the first column only.
# For time series, the first (index: 0) column must be sliced out here for the subsequent commands to work.

BEV_share_cT = df_BEVshare.values # convert pandas dataframe values to numpy array
```

```python
I_c.shape # inspect the shape of each array or time series
```

```python
BEV_share_cT.shape # inspect the shape of each array or time series
```

```python
I_cT = np.einsum('c,cT->cT',I_c,BEV_share_cT[0:35,:]) / 100 
# Here, we pass two arrays, I_c, and BEV_share_cT[0:35,:], to the einsum function.
# In the formatting string, we specify the shape of the two: c and cT, and specify the shape of the final result: ->cT.
# More on np.einsum in the subsequent software workbooks!
```

In the command above, the first 35 rows (age-cohorts) are sliced out so that the age-cohort range of _BEVshare_ matches the one of _I_, which covers the years 1990-2024.

We plot the result:

```python
plt.plot(np.arange(1990,2025,1),I_cT) # plot with given x-axis values, generated with the np.arange command
plt.ylim(0, 1.1*np.max(I_cT)) # set y axis maximum
plt.xlabel('age-cohort') # add x axis label 
plt.ylabel('new registration (inflow), cars') # add y axis label
plt.title('Inflow of pass. cars by age-cohort and drive technology, EUR 20 countries') # add title
plt.legend(['gasoline vehicles','battery electric vehicles']) # add a legend
plt.savefig('Inflow_EUR20_cT.png', dpi = 300) # save figure with the modifications made above. No plt.show() command when saving the figure!
```

We see that electric vehicles are on the rise in the EUR countries, with the following values and total:

The largest new registration of electric vehicles so far in the study region was about 2.05 million/year (in 2023), and the cumulative total so far was about 9.1 million units.

**Export the inflow time series by technology to excel**

First, we create new dataframe from the calculated result and store it to Excel for archiving and futher analysis/inspection.

```python
export_df= pd.DataFrame(data=I_cT,      # values
            index=np.arange(1990,2025,1),        # 1st column as index
            columns=['gasoline vehicles','battery electric vehicles'])  # 1st row as the column names
export_df
```

```python
export_df.to_excel('Bath_dMFA_2026_Software_1_Output_Single.xlsx', merge_cells=False)
```

Second, we create another dataframe and export several dataframes to different Excel sheets, using an xlsxwriter instance.

```python
pd_xlsx_writer = pd.ExcelWriter('Bath_dMFA_2026_Software_1_Output_Multiple.xlsx')
```

```python
ExportData2 = np.cumsum(I_cT, axis = 0) # cumulative sum over all inflow until current year
ExportData2.shape
```

```python
export_df.to_excel(pd_xlsx_writer, sheet_name="Inflow_cT", merge_cells=False)   

export_df2= pd.DataFrame(data=ExportData2,      # values: sum up I_cT over first axis (index 0: age-cohort)
            index=np.arange(1990,2025,1),        # 1st column as index
            columns=['gasoline vehicles','battery electric vehicles'])   # 1st row as the column names

export_df2.to_excel(pd_xlsx_writer, sheet_name="Inflow_Cumulative_T", merge_cells=False)   

pd_xlsx_writer.close()

```
