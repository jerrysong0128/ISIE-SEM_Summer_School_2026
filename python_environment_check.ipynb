## 1. Which Python is running?

`sys.executable` is the single most useful diagnostic: it is the full path to the Python interpreter executing this notebook.

```python
import sys

print("Python executable (sys.executable):")
print("   ", sys.executable)
print()
print("Python version:")
print("   ", sys.version.replace("\n", " "))
```

## 2. Is this the *right* environment?

For a **uv** project the path above should contain `.venv` (e.g. `.../mfa-env-test/.venv/...`).
For a **Miniforge/conda** environment it should contain `envs` or `miniforge`.

If it instead points at a system-wide Python (`/usr/bin/python`, `C:\\Python3x\\...`,
the Windows Store path, or a bare `base` conda env), the notebook is **not** using your
project environment and the package checks below will probably fail.

```python
import sys

exe = sys.executable.replace("\\", "/")
print("Environment root (sys.prefix):")
print("   ", sys.prefix)
print()

if ".venv" in exe:
    print("Looks like a uv / virtual environment (.venv).  \u2705")
elif "miniforge" in exe or "/envs/" in exe or "conda" in exe:
    print("Looks like a conda / Miniforge environment.  \u2705")
else:
    print("Could not confidently identify a project environment.  \u26a0\ufe0f")
    print("If this is a system or base Python, you are probably not in your project env.")
```

## 3. Where does Python look for packages?

`sys.path` is the ordered list of directories Python searches when you `import` something. You should see a `site-packages` directory *inside* your environment near the top.

```python
import sys

print("sys.path (import search order):")
for p in sys.path:
    print("   ", p or "(current working directory)")
```

## 4. What is installed in this environment?

This lists every distribution visible to *this* interpreter, with versions — the same information `uv pip list` or `pip list` would show, read straight from the running environment.

```python
from importlib.metadata import distributions

pkgs = sorted(
    ((d.metadata["Name"], d.version) for d in distributions()),
    key=lambda x: x[0].lower(),
)
print(f"{len(pkgs)} packages installed in this environment:\n")
for name, ver in pkgs:
    print(f"   {name}=={ver}")
```

## 5. Are the course packages importable?

Here are some common numerical Python packages, but depending on what you're working on you may want to adjust this to reflect the packages you actually need!

```python
import importlib

# >>> EDIT THIS LIST to match your dependencies <<<
to_check = ["numpy", "pandas", "matplotlib", "scipy", "openpyxl"]

print("Importing key packages:\n")
all_ok = True
for name in to_check:
    try:
        mod = importlib.import_module(name)
        ver = getattr(mod, "__version__", "unknown")
        print(f"   \u2705 {name:<12} {ver}")
    except ImportError as exc:
        all_ok = False
        print(f"   \u274c {name:<12} NOT INSTALLED ({exc})")

print()
print("All key packages present. \U0001f389" if all_ok
      else "Some packages are missing \u2014 see \u274c above.")
```

## 6. Does plotting work inline?

If you see a sine/cosine curve below the cell, your notebook front-end and matplotlib are wired up correctly.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 4 * np.pi, 200)
fig, ax = plt.subplots(figsize=(6, 3))
ax.plot(x, np.sin(x), label="sin")
ax.plot(x, np.cos(x), label="cos")
ax.set_title("If you can see this curve, your notebook + matplotlib work")
ax.legend()
plt.show()
```
